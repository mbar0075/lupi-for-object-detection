### Evaluation Metrics by Model

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Dictionary of models and paths
plotting_dictionary = [
    {'title': "RetinaNet\n (RGB)", 'path': "../runs/RetinaNet/RetinaNet_rgb1"},
    {'title': "FCOS\n (RGB)", 'path': "../runs/FCOS/FCOS_rgb1"},
    {'title': "Faster R-CNN\n (RGB)", 'path': "../runs/FasterRCNN/FasterRCNN_rgb1"},
    {'title': "SSD\n (RGB)", 'path': "../runs/SSD/SSD_rgb1"},
    {'title': "SSDLite\n (RGB)", 'path': "../runs/SSDLite/SSDLite_rgb1"},
]

# Metrics to extract
metrics = ["map", "map_50", "map_75", "mar_1", "mar_10", "mar_100"]
metric_labels = {
    "map": "Mean Average Precision (mAP@50-95)",
    "map_50": "mAP@50",
    "map_75": "mAP@75",
    "mar_1": "Mean Average Recall (mAR@1)",
    "mar_10": "mAR@10",
    "mar_100": "mAR@100"
}

data = []
labels = []

# Read data
for entry in plotting_dictionary:
    file_path = os.path.join(entry['path'], "coco_mAP.csv")
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        values = [df.iloc[0][metric] for metric in metrics]
        data.append(values)
        labels.append(entry['title'])
    else:
        print(f"Warning: {file_path} not found")
        data.append([None] * len(metrics))

# Convert to NumPy array
data = np.array(data, dtype=float)

# Set Seaborn style
sns.set_style("darkgrid")
colors = sns.color_palette("hls", len(metrics))  # Better bar colors

# Plotting
x = np.arange(len(labels))
width = 0.15  # Increased width

fig, ax = plt.subplots(figsize=(20, 10))
for i, (metric, color) in enumerate(zip(metrics, colors)):
    bars = ax.bar(x + i * width, data[:, i], width, label=metric_labels[metric], color=color, edgecolor='black')
    
    # Add numbers on top of bars
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval, f"{yval:.2f}", ha='center', va='bottom', fontsize=12)

ax.set_xlabel("Models", fontsize=18)
ax.set_ylabel("Metric Value", fontsize=18)
ax.set_title("Comparison of Evaluation Metrics Across Models (Prediction as Multiple Types of Litter)", fontsize=20)
ax.set_xticks(x + width * (len(metrics) / 2))
ax.set_xticklabels(labels, fontsize=14)
ax.legend(fontsize=14)
plt.tight_layout()
plt.show()
```

### Evaluation Metrics by Class

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Dictionary of models and paths
plotting_dictionary = [
    {'title': "RetinaNet\n (RGB)", 'path': "../runs/RetinaNet/RetinaNet_rgb1"},
    {'title': "RetinaNet\n (RGB + 3 Channels)", 'path': "../runs/RetinaNet/RetinaNet_rgb_itti_DA_HoG1"},
    {'title': "FCOS\n (RGB)", 'path': "../runs/FCOS/FCOS_rgb1"},
    {'title': "FCOS\n (RGB + 3 Channels)", 'path': "../runs/FCOS/FCOS_rgb_itti_DA_HoG1"},
    {'title': "Faster R-CNN\n (RGB)", 'path': "../runs/FasterRCNN/FasterRCNN_rgb1"},
    {'title': "Faster R-CNN\n (RGB + 3 Channels)", 'path': "../runs/FasterRCNN/FasterRCNN_rgb_itti_DA_HoG1"},
    {'title': "SSD\n (RGB)", 'path': "../runs/SSD/SSD_rgb1"},
    {'title': "SSD\n (RGB + 3 Channels)", 'path': "../runs/SSD/SSD_rgb_itti_DA_HoG1"},
    {'title': "SSDLite\n (RGB)", 'path': "../runs/SSDLite/SSDLite_rgb1"},
    {'title': "SSDLite\n (RGB + 3 Channels)", 'path': "../runs/SSDLite/SSDLite_rgb_itti_DA_HoG1"},
]

# Lists to store extracted data
data = {}
classes = None  # To store class labels

# Read data
for entry in plotting_dictionary:
    file_path = os.path.join(entry['path'], "coco_mAP.csv")
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        
        # Extract `map_per_class` and `mar_100_per_class`
        map_per_class = np.fromstring(df.iloc[0]['map_per_class'].strip('[]'), sep=' ')
        mar_100_per_class = np.fromstring(df.iloc[0]['mar_100_per_class'].strip('[]'), sep=' ')
        
        if classes is None:
            classes = np.fromstring(df.iloc[0]['classes'].strip('[]'), sep=' ', dtype=int)
        
        data[entry['title']] = (map_per_class, mar_100_per_class)
    else:
        print(f"Warning: {file_path} not found")

# Set Seaborn style
sns.set_style("darkgrid")
colors = sns.color_palette("plasma", len(classes))  # Unique color per class
class_names = ['Background', 'Clear Plastic Bottle', 'Drink Can', 'Drink Carton', 'Glass Bottle', 'Glass Jar', 'Other Plastic Bottle']

# Plotting
fig, ax = plt.subplots(figsize=(20, 10))
width = 0.15  # Bar width
x = np.arange(len(plotting_dictionary))  # X positions

for i, class_id in enumerate(classes):
    map_values = [data[model][0][i] if model in data else 0 for model in data]
    bars = ax.bar(x + i * width, map_values, width, label=class_names[class_id] + ' (AP@50-95)', color=colors[i], edgecolor='black')
    
    # Add values on top of bars
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval, f"{yval:.2f}", ha='center', va='bottom', fontsize=8)

ax.set_xlabel("Models", fontsize=18)
ax.set_ylabel("Average Precision per Class (0.5-0.95)", fontsize=18)
ax.set_title("Comparison of Average Precision per Class Across Models (Prediction as Multiple Types of Litter)", fontsize=20)
ax.set_xticks(x + width * (len(classes) / 2))
ax.set_xticklabels(data.keys(), fontsize=14)
ax.legend(fontsize=14, title="Classes")
plt.tight_layout()
plt.show()
```

### Evaluation Metrics for Privileged Information Generation

```python
import os
import seaborn as sns
import matplotlib.pyplot as plt

# Number of images
no_of_images = 452
path = "../Privileged Information/results"

duration_data = {}

for file_name in os.listdir(path):
    if file_name.endswith(".txt"):
        model_name = file_name.replace(".txt", "").replace("_", " ").title()
        file_path = os.path.join(path, file_name)
        
        with open(file_path, "r") as f:
            line = f.readline().strip()
            if "Duration:" in line:
                duration = float(line.split()[1])
                duration_per_image = duration / no_of_images
                duration_data[model_name] = duration_per_image

# Function to insert \n after every two spaces
def format_model_name(name):
    parts = name.split(" ")
    formatted_name = []
    for i, part in enumerate(parts):
        formatted_name.append(part)
        if (i + 1) % 2 == 0 and i + 1 < len(parts):  # Every 2 words
            formatted_name.append("\n")
    return " ".join(formatted_name)

# Modify Duration Data keys
duration_data = {format_model_name(key): value for key, value in duration_data.items()}

# Set Seaborn style
sns.set_style("darkgrid")
colors = sns.color_palette("hls", len(duration_data))

# Plot timing data
fig, ax = plt.subplots(figsize=(15, 7))
ax.bar(duration_data.keys(), duration_data.values(), color=colors, edgecolor='black')

# Add values on top of bars
for key, value in duration_data.items():
    ax.text(key, value, f"{value:.2f} s/img", ha='center', va='bottom', fontsize=10)

ax.set_xlabel("Models", fontsize=14)
ax.set_ylabel("Seconds per Image", fontsize=14)
ax.set_title("Time Taken to Generate Privileged Information", fontsize=16)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
